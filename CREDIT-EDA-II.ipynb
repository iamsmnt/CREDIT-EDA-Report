```python
def univariate_pie(variable):
    
    values_1 = defaulter_application[variable].value_counts().values
    labels_1 = defaulter_application[variable].value_counts().index
    
    values_0 = non_defaulter_application[variable].value_counts().values
    labels_0 = non_defaulter_application[variable].value_counts().index
    
    color = sns.color_palette("pastel", 5)
    fig, ax = plt.subplots(1, 2,figsize=(14,8))
    ax[0].pie(values_1, labels = labels_1,explode=(0,0.02), autopct = '%0.0f%%', colors = color)
    ax[0].set_ylabel('Total Counts')
    ax[0].set_title(f'Distribution of {variable} for Defaulters',fontsize=15)
    ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=40, ha="right")
    
    ax[1].pie(values_0, labels = labels_0,explode=(0,0.02), autopct = '%0.0f%%', colors = color)
    ax[1].set_ylabel('Total Counts')
    ax[1].set_title(f'Distribution of {variable} for Non-Defaulters',fontsize=15)
    ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=40, ha="right")

    
    plt.show()
```

```python
def univariate_bar(variable):
    fig,(axes1,axes2) = plt.subplots(1,2,figsize=(20,6))
    
    sns.countplot(x=variable, data=defaulter_application,ax=axes1)
    axes1.set_ylabel('Total Counts')
    axes1.set_title(f'Distribution of {variable} for Defaulters',fontsize=15)
    axes1.set_xticklabels(axes1.get_xticklabels(), rotation=40, ha="right")
    
    #normalized Percentages
    for p in axes1.patches:
        axes1.annotate('{:.1f}%'.format((p.get_height()/len(defaulter_application))*100), (p.get_x()+0.1, p.get_height()+50))
        
    sns.countplot(x=variable, data=non_defaulter_application,ax=axes2)
    axes2.set_ylabel('Total Counts')
    axes2.set_title(f'Distribution of {variable} for Non-Defaulters',fontsize=15)    
    axes2.set_xticklabels(axes2.get_xticklabels(), rotation=40, ha="right")
    
    #normalized Percentages
    for p in axes2.patches:
        axes2.annotate('{:.1f}%'.format((p.get_height()/len(non_defaulter_application))*100), (p.get_x()+0.1, p.get_height()+50))

    
    plt.show()
```
